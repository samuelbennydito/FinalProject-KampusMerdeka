### Pendahuluan

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df = pd.read_excel("kesan - boost career.xlsx")
```

```python
df.head()
```

```python
df.info()
```

### Data Cleaning

```python
df.isna().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.info()
```

```python
df.astype("str")
```

### Data Visualization

```python
word_cloud = WordCloud(collocations = False, background_color = 'white').generate(str(df))
```

```python
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# show only 10 words in the wordcloud .
wordcloud = WordCloud(width=480, height=480, max_words=10).generate(str(df))
```

```python
# plot the WordCloud image
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
